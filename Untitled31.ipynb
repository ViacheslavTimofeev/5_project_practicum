```python
import pandas as pd
import matplotlib.pyplot as plt

visits = pd.read_csv('D:\isitors.csv', sep='\t')
visits['date_time'] = pd.to_datetime(visits['date_time'], format='%Y.%m.%d %H:%M:%S')
visits['local_time'] = visits['date_time'] + pd.Timedelta(hours=3)
visits['date_hour'] = visits['local_time'].dt.round('1H')
#print(visits['date_hour'])
vis_pivot = visits.pivot_table(index='name', values='time_spent')
total_visits = len(visits) # 317104
total_stations = len(visits['id'].unique())
unique_count = len(visits['name'].unique()) # 45
total_days = 7
station_visits_per_day = total_visits / total_stations / total_days
visits = visits.sort_values(by='time_spent', ascending=False)
#print(visits.columns)
#sample = visits.query('id == "3c1e4c52"')
#sample.plot(x='local_time', y='time_spent', ylim=(0, 1000), grid=True, style='o', figsize=(12, 6))

visits['too_fast'] = visits['time_spent'] < 60
visits['too_slow'] = visits['time_spent'] > 1000

too_fast_stat = visits.pivot_table(index='id', values='too_fast')
good_ids = too_fast_stat.query('too_fast < 0.5')
good_data = visits.query('id in @good_ids.index and 60 <= time_spent <= 1000')
station_stat = visits.pivot_table(index='id', values='time_spent', aggfunc='median')
good_stations_stat = good_data.pivot_table(index='id', values='time_spent', aggfunc='median')
good_stat = good_data.pivot_table(index='name', values='time_spent', aggfunc='median')
stat = visits.pivot_table(index='name', values='time_spent')

stat['good_time_spent'] = good_stat['time_spent']
id_name = good_data.pivot_table(index='id', values='name', aggfunc=['count', 'first'])
id_name.columns = ['count', 'name']

station_stat_full = id_name.join(good_stations_stat)
good_stat2 = (
    station_stat_full
    .query('count > 30')
    .pivot_table(index='name', values='time_spent', aggfunc=['median', 'count'])
)
good_stat2.columns = ['median_time', 'stations']
final_stat = stat.join(good_stat2)
#station_stat_full.plot(x='count', y='time_spent', kind='scatter', grid=True)
station_stat_multi = visits.pivot_table(index='id', values=['time_spent', 'too_fast', 'too_slow'])
#pd.plotting.scatter_matrix(station_stat_multi, figsize=(9, 9))
station_stat_multi['good_time_spent'] = good_stations_stat['time_spent']
#print(station_stat_multi.corr())
#pd.plotting.scatter_matrix(station_stat_multi, figsize=(9, 9))
final_stat = final_stat.dropna(subset=['median_time'])
final_stat = final_stat.sort_values(by='median_time')
big_nets_stat = final_stat.query('stations > 10')

station_stat_full['group_name'] = (
station_stat_full['name']
.where(station_stat_full['name'].isin(big_nets_stat.index), '')
)
stat_grouped = (
     station_stat_full
    .query('count > 30')
    .pivot_table(index='group_name', values='time_spent', aggfunc=['median', 'count'])
)
stat_grouped.columns = ['time_spent', 'count']
stat_grouped = stat_grouped.sort_values(by='time_spent')

good_data['group_name'] = (
good_data['name']
.where(good_data['name'].isin(big_nets_stat.index), '')
)

for name, group_data in good_data.groupby('group_name'):
    group_data.plot(kind='hist', y='time_spent', title=name,  bins=50)

     
#stat_grouped.plot(y='count', kind='pie', figsize=(8, 8))
#print(stat_grouped)
#print(big_nets_stat)
#final_stat.plot(kind='bar', figsize=(10, 5), grid=True)
#final_stat.hist(bins=100)
#print(station_stat_full['count'].corr(station_stat_full['time_spent']))
#print(final_stat)
#print(station_stat_full)
#station_stat_full.hist(bins=30)
#station_stat_full.hist(bins=30, range=(0, 300))
#print(visits['too_fast'].mean()) # 0.42213910893586964
#print(visits['too_fast'])
#print(sample)
#visits.hist(bins=100, range=(0, 1500))
#plt.show()
#sample.hist(bins=100, range=(0, 1500))
#plt.show()
```
